```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

```python
def slope(t, y):
    return 3*t**2/(2-y)
```

```python
y0 = np.array([0, 0])
t0 = 0
tf = 10
ts = (t0, tf)
n = 1000
T = np.linspace(t0, tf, n)
```

```python
sol = solve_ivp(slope, ts, y0, method='BDF', t_eval=T)
```

```python
plt.xlabel("t")
plt.ylabel("y")
plt.title("What happened?")
plt.plot(sol.t, sol.y[0])
plt.show()
```
